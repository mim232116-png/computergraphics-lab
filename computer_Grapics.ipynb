```python
# DDA Line Drawing Algorithm with Plotting
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    points = []  # to store the generated pixel points

    dx = x2 - x1
    dy = y2 - y1

    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)

    Xinc = dx / steps
    Yinc = dy / steps

    x, y = x1, y1
    for i in range(int(steps) + 1):
        points.append((round(x), round(y)))
        x += Xinc
        y += Yinc

    return points

# Example usage
x1, y1 = 2, 3
x2, y2 = 15, 10

points = dda_line(x1, y1, x2, y2)

# Plotting
xs, ys = zip(*points)
plt.figure(figsize=(6,6))
plt.plot(xs, ys, marker="s", color="blue", markersize=8, linestyle="None", label="DDA Pixels")
plt.plot([x1, x2], [y1, y2], color="red", linestyle="--", label="Actual Line")
plt.grid(True)
plt.legend()
plt.title("DDA Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Print the generated points
print("Generated Points:", points)
```

```python
# Midpoint Line Drawing Algorithm (all slopes) with Plotting
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    points = []

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1

    # Decision parameter
    if dx > dy:
        d = 2*dy - dx
        x, y = x1, y1
        for _ in range(dx + 1):
            points.append((x, y))
            if d >= 0:
                y += sy
                d -= 2*dx
            x += sx
            d += 2*dy
    else:
        d = 2*dx - dy
        x, y = x1, y1
        for _ in range(dy + 1):
            points.append((x, y))
            if d >= 0:
                x += sx
                d -= 2*dy
            y += sy
            d += 2*dx

    return points

# Example usage
x1, y1 = 2, 3
x2, y2 = 15, 10

points = midpoint_line(x1, y1, x2, y2)

# Plotting
xs, ys = zip(*points)
plt.figure(figsize=(6,6))
plt.plot(xs, ys, marker="s", color="red", markersize=8, linestyle="None", label="Midpoint Pixels")
plt.plot([x1, x2], [y1, y2], color="green", linestyle="--", label="Actual Line")
plt.grid(False)
plt.legend()
plt.title("Mid-Point Line Drawing Algorithm (All Slopes)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Print generated points
print("Generated Points:", points)
```

```python
# Bresenham Line Drawing Algorithm with Plotting
import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    points = []

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1

    err = dx - dy
    x, y = x1, y1

    while True:
        points.append((x, y))
        if x == x2 and y == y2:
            break
        e2 = 2 * err
        if e2 > -dy:
            err -= dy
            x += sx
        if e2 < dx:
            err += dx
            y += sy

    return points


# ===== Example =====
x1, y1 = 2, 3
x2, y2 = 15, 10

points = bresenham_line(x1, y1, x2, y2)

# Plotting
xs, ys = zip(*points)
plt.figure(figsize=(6,6))
plt.plot(xs, ys, marker="s", color="blue", markersize=8, linestyle="None", label="Bresenham Pixels")
plt.plot([x1, x2], [y1, y2], color="red", linestyle="--", label="Ideal Line")
plt.grid(False)
plt.legend()
plt.title("Bresenham Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Print generated points
print("Generated Points:", points)
```

```python
# Midpoint Circle Drawing Algorithm with Plotting
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r):
    points = []
    x = 0
    y = r
    d = 1 - r  # decision parameter

    while x <= y:
        # 8-way symmetry
        points.extend([
            ( xc + x, yc + y),
            ( xc - x, yc + y),
            ( xc + x, yc - y),
            ( xc - x, yc - y),
            ( xc + y, yc + x),
            ( xc - y, yc + x),
            ( xc + y, yc - x),
            ( xc - y, yc - x)
        ])

        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        x += 1

    return points


# ===== Example =====
xc, yc = 10, 10   # center
r = 8             # radius

points = midpoint_circle(xc, yc, r)

# Plotting
xs, ys = zip(*points)
plt.figure(figsize=(6,6))
plt.plot(xs, ys, marker="s", color="blue", markersize=6, linestyle="None", label="Circle Pixels")

# Draw ideal circle for comparison
circle = plt.Circle((xc, yc), r, color="red", fill=False, linestyle="--", label="Ideal Circle")
plt.gca().add_artist(circle)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(False)
plt.legend()
plt.title("Midpoint Circle Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Print generated points
print("Generated Points:", points)
```

```python
# 2D Transformation - Translation with Plotting
import matplotlib.pyplot as plt
import numpy as np

def translate(points, tx, ty):
    """Translate a set of 2D points by tx, ty."""
    translation_matrix = np.array([[1, 0, tx],
                                   [0, 1, ty],
                                   [0, 0, 1]])

    # Convert points to homogeneous coordinates
    homo_points = np.hstack((points, np.ones((points.shape[0], 1))))

    # Apply transformation
    translated_points = homo_points @ translation_matrix.T
    return translated_points[:, :2]


# Example shape (triangle)
points = np.array([[2, 2],
                   [6, 2],
                   [4, 6],
                   [2, 2]])   # closing the triangle

tx, ty = 4, 3   # translation values
translated_points = translate(points, tx, ty)

# Plot original vs translated shape
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")
plt.plot(translated_points[:,0], translated_points[:,1], 'ro--', label="Translated Shape")

plt.grid(True)
plt.legend()
plt.title(f"2D Translation (tx={tx}, ty={ty})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```

```python
# 2D Transformation - Scaling with Plotting
import matplotlib.pyplot as plt
import numpy as np

def scale(points, sx, sy):
    """Scale a set of 2D points by sx, sy."""
    scaling_matrix = np.array([[sx, 0, 0],
                               [0, sy, 0],
                               [0, 0,  1]])

    # Convert points to homogeneous coordinates
    homo_points = np.hstack((points, np.ones((points.shape[0], 1))))

    # Apply transformation
    scaled_points = homo_points @ scaling_matrix.T
    return scaled_points[:, :2]


# Example shape (triangle)
points = np.array([[2, 2],
                   [6, 2],
                   [4, 6],
                   [2, 2]])   # closing the triangle

sx, sy = 2, 1.5   # scaling factors
scaled_points = scale(points, sx, sy)

# Plot original vs scaled shape
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")
plt.plot(scaled_points[:,0], scaled_points[:,1], 'ro--', label="Scaled Shape")

plt.grid(True)
plt.legend()
plt.title(f"2D Scaling (sx={sx}, sy={sy})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```

```python
# 2D Transformation - Shearing with Plotting
import matplotlib.pyplot as plt
import numpy as np

def shear(points, shx=0, shy=0):
    """
    Shear a set of 2D points.
    shx: shear factor along X-axis
    shy: shear factor along Y-axis
    """
    shearing_matrix = np.array([[1, shx, 0],
                                [shy, 1, 0],
                                [0,  0, 1]])

    # Convert points to homogeneous coordinates
    homo_points = np.hstack((points, np.ones((points.shape[0], 1))))

    # Apply transformation
    sheared_points = homo_points @ shearing_matrix.T
    return sheared_points[:, :2]


# Example shape (triangle)
points = np.array([[2, 2],
                   [6, 2],
                   [4, 6],
                   [2, 2]])   # closing the triangle

shx, shy = 1, 0.5   # shear factors
sheared_points = shear(points, shx, shy)

# Plot original vs sheared shape
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")
plt.plot(sheared_points[:,0], sheared_points[:,1], 'ro--', label="Sheared Shape")

plt.grid(True)
plt.legend()
plt.title(f"2D Shearing (shx={shx}, shy={shy})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```

```python
# 2D Transformation - Rotation with Plotting
import matplotlib.pyplot as plt
import numpy as np

def rotate(points, angle_deg):
    """
    Rotate a set of 2D points about the origin.
    angle_deg: rotation angle in degrees (counter-clockwise)
    """
    angle_rad = np.radians(angle_deg)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad), 0],
                                [np.sin(angle_rad),  np.cos(angle_rad), 0],
                                [0, 0, 1]])

    # Convert points to homogeneous coordinates
    homo_points = np.hstack((points, np.ones((points.shape[0], 1))))

    # Apply transformation
    rotated_points = homo_points @ rotation_matrix.T
    return rotated_points[:, :2]


# Example shape (triangle)
points = np.array([[2, 2],
                   [6, 2],
                   [4, 6],
                   [2, 2]])   # closing the triangle

angle = 45   # rotation angle in degrees
rotated_points = rotate(points, angle)

# Plot original vs rotated shape
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")
plt.plot(rotated_points[:,0], rotated_points[:,1], 'ro--', label=f"Rotated Shape ({angle}°)")

plt.grid(True)
plt.legend()
plt.title(f"2D Rotation by {angle}°")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```

```python
# 2D Transformation - Reflection with Plotting
import matplotlib.pyplot as plt
import numpy as np

def reflect(points, axis='x'):
    """
    Reflect a set of 2D points.
    axis: 'x' -> reflect about X-axis
          'y' -> reflect about Y-axis
          'origin' -> reflect about origin
          'y=x' -> reflect about line y=x
    """
    if axis == 'x':
        reflection_matrix = np.array([[1, 0, 0],
                                      [0,-1, 0],
                                      [0, 0, 1]])
    elif axis == 'y':
        reflection_matrix = np.array([[-1, 0, 0],
                                      [0, 1, 0],
                                      [0, 0, 1]])
    elif axis == 'origin':
        reflection_matrix = np.array([[-1, 0, 0],
                                      [0,-1, 0],
                                      [0, 0, 1]])
    elif axis == 'y=x':
        reflection_matrix = np.array([[0, 1, 0],
                                      [1, 0, 0],
                                      [0, 0, 1]])
    else:
        raise ValueError("Invalid axis. Choose from 'x', 'y', 'origin', 'y=x'.")

    # Convert points to homogeneous coordinates
    homo_points = np.hstack((points, np.ones((points.shape[0], 1))))

    # Apply transformation
    reflected_points = homo_points @ reflection_matrix.T
    return reflected_points[:, :2]


# Example shape (triangle)
points = np.array([[2, 2],
                   [6, 2],
                   [4, 6],
                   [2, 2]])   # closing the triangle

axis = 'x'   # reflection axis
reflected_points = reflect(points, axis)

# Plot original vs reflected shape
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")
plt.plot(reflected_points[:,0], reflected_points[:,1], 'ro--', label=f"Reflected Shape (about {axis}-axis)")

plt.grid(True)
plt.legend()
plt.title(f"2D Reflection about {axis}-axis")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```
